```
rcparams = {"font.family":"serif", "mathtext.fontset":"stix"}
plt.rcParams.update(rcparams)
plt.rcParams["font.serif"] = ["Time New Roman"] + plt.rcParams["font.serif"]

mpl.rc('xtick', labelsize=7)
mpl.rc('ytick', labelsize=10)
```

```
Y_1 = np.load("/content/drive/My Drive/Colab Notebooks/result/memory/SD_Memory.npy") / 1000
Y_2 = np.load("/content/drive/My Drive/Colab Notebooks/result/memory/SC_Memory.npy") / 1000
Y_3 = np.load("/content/drive/My Drive/Colab Notebooks/result/memory/CNN_Memory.npy") / 1000
Y_4 = np.load("/content/drive/My Drive/Colab Notebooks/result/memory/FCN_Memory.npy") / 1000
```

```
list_size = range(4, 21, 2)
x_range = range(4, 21, 2)
x = range(4, 21, 2)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.plot(list_size, Y_1, '--', marker='s', markersize=4, color='darkorange', fillstyle='none', label='DUA-LFU')
plt.plot(list_size, Y_2, '-.', marker='d', markersize=5, color='green', fillstyle='none', label='CUA-LFU')
plt.plot(list_size, Y_4, color='r', marker='<', markersize=6, fillstyle='none', label='DQN-FCN')
plt.plot(list_size, Y_3, color='b', marker='o', markersize=5, fillstyle='none', label='Proposed scheme')

plt.xlabel("$M$, [chunks]", fontsize=13)
plt.xticks(x_range, x, rotation=0)
plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=10, top=True)
plt.ylabel("Average latency in a content delivery [ms]", fontsize=12)
plt.legend(fontsize=11, loc="best")
plt.grid(True, alpha=2.0, linestyle=':')
plt.show()
```
